# 03 · Word Embeddings & Distributional Meaning

Train Word2Vec on each corpus and *show* what BoW/TF-IDF (orthogonal one-hot bases) structurally cannot: semantic similarity.


```python
from nlp_system.data.make_dataset import load_split
from nlp_system.pipeline.preprocess import TextPreprocessor
from nlp_system.features.embeddings import train_word2vec, most_similar_report, plot_embeddings
from nlp_system.pipeline.nltk_setup import ensure_nltk
ensure_nltk()
```

## Tokenize the Amazon corpus


```python
df = load_split('amazon_fine_food', 'train')
pre = TextPreprocessor.for_reviews()
docs = [pre.tokens(t) for t in df['text']]
print('docs:', len(docs))
```

## Train Word2Vec


```python
model = train_word2vec(docs, vector_size=100, window=5, min_count=5, epochs=5)
print('vocab:', len(model.wv))
```

## Nearest neighbors

Sentiment-bearing words should pull in their synonyms — impossible for classical sparse vectors.


```python
from pprint import pprint
pprint(most_similar_report(model, ['great','awful','delicious','cheap','recommend']))
```

## 2-D projection


```python
path = plot_embeddings(model, n_words=200, method='tsne')
print('saved ->', path)
from IPython.display import Image; Image(str(path))
```

## Takeaway

Classical vectors place `excellent` and `great` in orthogonal dimensions (cosine 0). Embeddings place them adjacent. That distributional structure is the motivation for moving from TF-IDF features toward dense/transformer representations — the natural next step after this lab.
